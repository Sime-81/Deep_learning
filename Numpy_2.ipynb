```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
type(face)
face.shape
```

```python
PositionX = face.shape[0]
PositionY = face.shape[1]
print(PositionX, PositionY)
```

```python
print(int(PositionX/4), int(PositionY/4))
#Zoom 1/4 vers le milieux de l'image
```

```python
Middle = face[int((PositionX/2)- 97):int((PositionX/2)+ 96), int((PositionY/2)- 129):int((PositionY/2)+ 128)]
Middle
```

```python
plt.imshow(Middle, cmap=plt.cm.gray)
plt.show()
```

```python
Middle[Middle>200] = 255
Middle[Middle<55] = 0
Middle
```

```python
plt.imshow(Middle, cmap=plt.cm.gray)
plt.show()
```

```python
#Correction de l'exercice
X = face.shape[0]
Y = face.shape[1]
zoom_face = face[X//4:-X//4, Y//4:-Y//4]
zoom_face
```

```python
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```
