### Travailler sur plusieur Graphique avec Subplot

La fonction subplot va permettre de d'intgr plusieur graphique group ensemble 

dclaration :

plt.subplot([nombre de ligne], [nombre de colone], [Graphique sur lequelle on travail])

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, c='red')
plt.subplot(2, 1, 2)
plt.plot(x, y, c='blue')
```

##### Cycle de vie d'un graphique Subplot

Le cycle de vie de ce genre de graphique auront une structure particulire qu'il faudra respect

___

plt.figure() --> Dbut de la figure

plt.subplot(2, 1, 1) --> Dbut du travail avec subplot (exemple: ici nous travaillons sur un graphique  2 ligne et 1 colone)

___
Graphique 1


plt.plot(...) /

................ / --> Instruction concernant le graphique 1

................ /

___
plt.subplot(2, 1, 2) --> Changement de graphique sur lequelle on travail
___
Graphique 2


plt.plot(...) /

................ / --> Instruction concernant le graphique 2

................ /

___

plt.show() --> affichage de la figure

___




###### Dmonstration du cycle de vie :

```python
#Initialisation de l'espace de travail
plt.figure()
plt.subplot(2, 1, 1) #dclaration de travail sur le graphique 1 + dimension du groupement de graphique

#Initialisation du premier graphique 
plt.plot(x, y)
plt.title('Graphique 1')

#changement d'espace de travail
plt.subplot(2, 1, 2) #dclaration de travail sur le graphique 2

#Initialisation du deuxime graphique 
plt.plot(x, np.sin(x), label='sinus')
plt.plot(x, np.cos(x), label='cosinus')
plt.legend()

#affichage du graphique
plt.show()
```

## 2) OOP

Avec OOP nous pouvons faire la mme chose qu'avec les fonctions mais en plus poussez ceci peut tre utile mais n'est pas recommander si vous n'avez besoin que de graphique simple, surtout  cause du grand problme quad on travail sur plussieur graphique et que notre espace de travail devient un tableau ndarray numpy 

```python
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

Jusqu'ici aucun problme nous avons effectuer exactemen le mme graphique qu'avant avec nos fonction mais je vais maintenant vous montr ce qui se passerait si je crais plusieur graphique grouper comme on  dja vu au paravant

```python
fig, ax = plt.subplots(2, 1, sharex=True) #L'argument sharex permet de partager l'axe des x (pas important)
ax.plot(x, y)
ax.plot(x, np.sin(x))
plt.show()
```

Voici le problme dont je parlais plus haut nous rencontron cette erreur :

AttributeError: 'numpy.ndarray' object has no attribute 'plot'

mais me direz vous un tableau numpy mais de quoi il me parle  ce programme ?

ce quoi je vous rpondrais que c'est enfaite un particulariter de cette fonction qui pour stocker nos deux graphique va devenir un tableau numpy  ndimension 

```python
fig, ax = plt.subplots(2, 1, sharex=True) 
#ax.plot(x, y)
#ax.plot(x, np.sin(x))
#plt.show()

type(ax)
```

Ici nous voyons bien qu'enfaite notre variable [ax] est un tableau numpy ce qui explique donc l'erreur que nous recevons comme retour

comme nous travaillons maintenant avec un tableau numpy il faut adapter notre criture de code  un tableau numpy, rien de plus simple.

```python
fig, ax = plt.subplots(2, 1, sharex=True) 
ax[0].plot(x, y)
ax[1].plot(x, np.sin(x))
plt.show()
```

Et voyez donc plus aucune erreur. Voici le principale dfaut de cette mthode voil pourquoi je vous recommande de ne pas l'utiliser car ce genre de chose peut grandement vous compliquez la tache dans vos programme

## 3) Exercice

Afficher 4 Graphique de base de donner diffrente avec un dictionnaire qui comprend N datasets.

```python
dataset = {f"experience{i}": np.random.randn(100) for i in range(4)}
#dataset
```

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12, 12))
    
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
    
    plt.show()
```

```python
graphique(dataset)
```
