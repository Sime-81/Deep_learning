# MATPLOTLIB

```python
import numpy as np
```

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

## 1) Pyplot

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
```

```python
#argument [c] pour set la couleur de la courbe
plt.plot(x, y, c="red")
```

```python
#argument [lw] pour set l'psseur de la courbe
plt.plot(x, y, c="black", lw=3)
```

```python
#argument [ls]  pour set le style de la ligne
plt.plot(x, y, c="black", lw=3, ls="--")
```

### Cycle de vie Matplotlib 

Il est importan de comprendre qu'il ya un dbut  l'initialisation de ces graphique et bien entendu il faut respect une structure prcise et si tenir comme dans tout les programme qui se respect

les instrcutions commenceront toujours ainsi :

plt.figure()  --> Dbut de la figure (Permet de redimensionn la taille du tableau si besoin)

pl.plot(..., ...) /

.................... / --> Contenu (qui peut contenir toute les autres instructions sur le graphique)

.................... /

plt.show() --> pour afficher (on n'est pas oblig de le mettre si on travail avec Jupiter comme IDE)

plt.savefig('[Nom].[extension]') --> pour sauvegarder notre figure dans notre projet
_____

exemple :

```python
plt.figure()
```

```python
plt.plot(x, y)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x, y)
```

```python
plt.figure()
plt.plot(x, y)
plt.plot(x, x**3)
```

Nous pouvons ajouter beaucoup d'autre argument sur notre graphique 

exemple :

```python
plt.figure()
plt.plot(x, y, label='quadratique')
plt.plot(x, x**3, label='cubique')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend() #pour la lgende il utilise les paramtre label dans les plot()
plt.show()
plt.savefig('fig1.png')
```
