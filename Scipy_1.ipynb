## 1) Interpolation avec Interpolate

Lorsque nous retrouvons des dataset incomplet contenant la valeur NaN il est parfois difficile d'afficher correctement nos donnés, Interpolate va nous permettre de crée une fonction à partir de nos données existante et ajouter des donnés potientiellemnt plausible. Attention à ce que l'interpolation ne cache pas la réalité. Une mauvaise interpolation pourrait causer beaucoup de problème ou ne pas tenir compte de certaine donnés qui était considérer par cette interpolation comme superflue ou unique.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#création d'une base de donnés pour la démonstration
x = np.linspace (0, 10, 10)
y = x**2
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(x, y, kind='linear') # interp1d nous retourne une fonction f(x) = ...
# On utilisera une fonction linéaire pour commencer

new_x = np.linspace(0, 10, 30) # Création d'un nouveaux axe x en 30 point
result = f(new_x) # application de la fonction sur le nouvelle axe x

# Afffichage du resultat 
plt.scatter(x, y)
plt.scatter(new_x, result, c='r') 
```
