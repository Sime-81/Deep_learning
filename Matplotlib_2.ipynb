## (5) Graphique de Classification avec plt.scatter

ici nous utilisons une base de donnée sur les fleur de lisse représenter par x 4 colone de 150 ligne de donné et y de 150 données répartie entre 0 et 2. ici un graphique linéaire n'est pas très pertinent alors nous utiliseront un graphique scatter pour représenter nos donnée en nuage de point.

```python
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.7, s=70) #expliquation des différen paramètre si dessous
plt.xlabel('longueur sépale')
plt.ylabel('largeur sépale')
```

Le paramétre [c] fixer sur y nous permet de changer la couleur des point cela nous permet de classifier les point des donnée y {0, 1, 2} pour mieux visualiser. Le paramétre [Alpha] nous permet de controler la transparence des points. Le paramètre [s] nous permet de controler la taille des point.

## (4) Graphique 3D

Nous allons voir comment afficher un graphique en 3D ce qui nous permettra d'afficher la casi totaliter de notre base de donnée 

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax = plt.axes(projection='3d')
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y)
```

Pour voir la version 3D navigable vous puvez utiliser le script [3dGraphique(1).py] 

Ici nous pouvons voir un nuage de point en 3d. Malgré que nous devions utiliser la version orienter objet de matplotlib elle est très puissante et très utile mais voyons voire comment elle se comporte pour afficher des fonction mathèmatique 

```python
f = lambda x, y: np.sin(x) + np.cos(x + y)
```

```python
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
```

Ici nous avons modéliser un fonction en 3d je sais que les ligne de code paraisse complexe mais nous avons juste modéliser les ax pour permettre l'affichage de la fonction. l'argument [cmap] control la couleur de se graphuiqe et nous avons utiliser plot_surface pour ce rendu.

## (2) Contour plots

Ce type de graphique est super utile quand nous voulons visualiser un model qui occupe 3 dimension en vu du dessu. En intelligence artificielle nous cherchons souvent à maximiser une fonction, ce type de graphique vous permetra une meilleur compréhention de ce que vous essayer de maximiser.

Nous allons réutiliser la fonction que nous avons fait passer en 3D

```python
plt.contour(X, Y, Z, 20, colors='black')
```

L'argument 20 est utiliser pour la présition des couche au plus le nombre est grand au plus c'est précis. l'argument [colors] nous permets de changer la couleur du graphique pour adapter à la lisibiliter souhaiter.

Je vais écrire une fonction plus complexe qui vous permetra de mieux comprendre à quel point ce type de graphique est pratique (les ligne c'est positifs et les pointier négatif)

```python
f = lambda x, y: np.sin(x) + np.cos(x + y)*np.cos(x)

X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, 20, colors='black')
```

Nous pouvons utiliser un autre type de contour plot pour visualiser ce graphique de manière différente 

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

Ce graphique nous permet de comprendre quel sont les valeur de nos différent graphique de manière plus visuel

## Excercice

utiliser le code de l'excercice de la doc Matplotlib_1 pour afficher nos quatre variable du data set de l'iris sur un seul et même figure.

```python
datasets = {f"data_iris{i + 1}" : x[:, i] for i in range(4)}

def graphique(data):
    n = len(data)
    plt.figure(figsize=(12, 18))
    
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
    
    plt.show()
```

```python
graphique(datasets)
```
