# Matplotlib les graphique important

Dans nos code de machine learning nous pouvons avoir besoin de manipuler des data-sets très différent des un des autres ce qui veux aussi dire qu'il faut les afficher de manière différente voila pourquoi je vais vous siter les Graphique important de Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target
name = list(iris.target_names)

print(f"x contient {x.shape[0]} exeples et {x.shape[1]} variable")
print(f"il y a {np.unique(y).size} classes")
```

## (5) Graphique de Classification avec plt.scatter

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.xlabel('longueur sépale')
plt.ylabel('largeur sépale')
```
