## (5) Graphique de Classification avec plt.scatter

ici nous utilisons une base de donnée sur les fleur de lisse représenter par x 4 colone de 150 ligne de donné et y de 150 données répartie entre 0 et 2. ici un graphique linéaire n'est pas très pertinent alors nous utiliseront un graphique scatter pour représenter nos donnée en nuage de point.

```python
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.7, s=70) #expliquation des différen paramètre si dessous
plt.xlabel('longueur sépale')
plt.ylabel('largeur sépale')
```

Le paramétre [c] fixer sur y nous permet de changer la couleur des point cela nous permet de classifier les point des donnée y {0, 1, 2} pour mieux visualiser. Le paramétre [Alpha] nous permet de controler la transparence des points. Le paramètre [s] nous permet de controler la taille des point.
